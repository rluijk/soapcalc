```python
import pandas as pd
import numpy as np
```

```python
df_base_props = pd.read_csv('./resources/SoapQualities.csv')

df_iodine = pd.read_csv('./resources/iodine.csv', header=None) # done
df_ins = pd.read_csv('./resources/ins.csv', header=None) # done
df_lauric = pd.read_csv('./resources/Lauric.csv', header=None) # done 
df_linoleic = pd.read_csv('./resources/linoleic.csv', header=None) # done
df_linolenic = pd.read_csv('./resources/linolenic.csv', header=None) # done
df_myristic = pd.read_csv('./resources/Myristic.csv', header=None) # done
df_oleic = pd.read_csv('./resources/Oleic.csv', header=None) # done
df_palmitic = pd.read_csv('./resources/palmitic.csv', header=None) # done
df_ricinoleic = pd.read_csv('./resources/Ricinoleic.csv', header=None)# done
df_SAP = pd.read_csv('./resources/SAP.csv', header=None)
df_stearic = pd.read_csv('./resources/Stearic.csv', header=None) # done
```

```python
df_base_props = df_base_props.apply(lambda x: x.astype(str).str.lower())
df_base_props.set_index('Oil', inplace=True)
df_base_props.head()
```

```python
df_SAP.dropna(axis = 0, how = 'all', inplace = True)

df_SAP.rename(columns={ df_SAP.columns[0]: "S" }, inplace = True)
df_SAP.rename(columns={ df_SAP.columns[1]: "A" }, inplace = True)
df_SAP.rename(columns={ df_SAP.columns[2]: "Oil" }, inplace = True)

df_SAP.drop(df_SAP.columns[3], axis=1, inplace=True)
df_SAP.drop(df_SAP.columns[3], axis=1, inplace=True)

df_SAP = df_SAP.apply(lambda x: x.astype(str).str.lower())
df_SAP.set_index('Oil', inplace=True)
df_SAP.head()
```

```python
df_stearic.dropna(axis = 0, how = 'all', inplace = True)
df_stearic.rename(columns={ df_stearic.columns[0]: "stearic value" }, inplace = True)
df_stearic.rename(columns={ df_stearic.columns[1]: "Oil" }, inplace = True)
df_stearic = df_stearic.apply(lambda x: x.astype(str).str.lower())
df_stearic.set_index('Oil', inplace=True)
df_stearic.head()
```

```python
df_ricinoleic.dropna(axis = 0, how = 'all', inplace = True)
df_ricinoleic.rename(columns={ df_ricinoleic.columns[0]: "ricinoleic value" }, inplace = True)
df_ricinoleic.rename(columns={ df_ricinoleic.columns[1]: "Oil" }, inplace = True)
df_ricinoleic = df_ricinoleic.apply(lambda x: x.astype(str).str.lower())
df_ricinoleic.set_index('Oil', inplace=True)
df_ricinoleic.head()
```

```python
df_palmitic.dropna(axis = 0, how = 'all', inplace = True)
df_palmitic.rename(columns={ df_palmitic.columns[0]: "palmitic value" }, inplace = True)
df_palmitic.rename(columns={ df_palmitic.columns[1]: "Oil" }, inplace = True)
df_palmitic = df_palmitic.apply(lambda x: x.astype(str).str.lower())
df_palmitic.set_index('Oil', inplace=True)
df_palmitic.head()
```

```python
df_oleic.dropna(axis = 0, how = 'all', inplace = True)
df_oleic.rename(columns={ df_oleic.columns[0]: "oleic value" }, inplace = True)
df_oleic.rename(columns={ df_oleic.columns[1]: "Oil" }, inplace = True)
df_oleic = df_oleic.apply(lambda x: x.astype(str).str.lower())
df_oleic.set_index('Oil', inplace=True)
df_oleic.head()
```

```python
df_myristic.drop(df_myristic.columns[2], axis=1, inplace=True)
df_myristic.dropna(axis = 0, how = 'all', inplace = True)
df_myristic.rename(columns={ df_myristic.columns[0]: "myristic value" }, inplace = True)
df_myristic.rename(columns={ df_myristic.columns[1]: "Oil" }, inplace = True)
df_myristic = df_myristic.apply(lambda x: x.astype(str).str.lower())
df_myristic.set_index('Oil', inplace=True)
df_myristic.head()
```

```python
df_linolenic.dropna(axis = 0, how = 'all', inplace = True)
df_linolenic.rename(columns={ df_linolenic.columns[0]: "linolenic value" }, inplace = True)
df_linolenic.rename(columns={ df_linolenic.columns[1]: "Oil" }, inplace = True)
df_linolenic = df_linolenic.apply(lambda x: x.astype(str).str.lower())
df_linolenic.set_index('Oil', inplace=True)
df_linolenic.head()
```

```python
df_linoleic.dropna(axis = 0, how = 'all', inplace = True)
df_linoleic.rename(columns={ df_linoleic.columns[0]: "linoleic value" }, inplace = True)
df_linoleic.rename(columns={ df_linoleic.columns[1]: "Oil" }, inplace = True)
df_linoleic = df_linoleic.apply(lambda x: x.astype(str).str.lower())
df_linoleic.set_index('Oil', inplace=True)
df_linoleic.head()
```

```python
df_lauric.dropna(axis = 0, how = 'all', inplace = True)
df_lauric.rename(columns={ df_lauric.columns[0]: "lauric value" }, inplace = True)
df_lauric.rename(columns={ df_lauric.columns[1]: "Oil" }, inplace = True)
df_lauric = df_lauric.apply(lambda x: x.astype(str).str.lower())
df_lauric.set_index('Oil', inplace=True)
df_lauric.head()
```

```python
df_iodine.drop(df_iodine.columns[2], axis=1, inplace=True)
df_iodine.dropna(axis = 0, how = 'all', inplace = True)
df_iodine.rename(columns={ df_iodine.columns[0]: "iodine value" }, inplace = True)
df_iodine.rename(columns={ df_iodine.columns[1]: "Oil" }, inplace = True)
df_iodine = df_iodine.apply(lambda x: x.astype(str).str.lower())
df_iodine.set_index('Oil', inplace=True)
df_iodine.head()
```

```python
df_ins.rename(columns={ df_ins.columns[0]: "INS value" }, inplace = True)
df_ins.rename(columns={ df_ins.columns[1]: "Oil" }, inplace = True)
df_ins.drop(df_ins.columns[2], axis=1, inplace=True)
df_ins.dropna(axis = 0, how = 'all', inplace = True)
df_ins = df_ins.apply(lambda x: x.astype(str).str.lower())
df_ins.set_index('Oil', inplace=True)
df_ins.head()
```

```python
from functools import reduce

#define list of DataFrames
dfs = [df_base_props, df_iodine, df_ins,df_lauric,df_linoleic,df_linolenic,df_myristic,df_oleic,
       df_palmitic,df_ricinoleic,df_SAP,df_stearic]

#merge all DataFrames into one
final_df = reduce(lambda  left,right: pd.merge(left,right,on=['Oil'],
                                            how='outer'), dfs)
```

```python

import os  
os.makedirs('output/test', exist_ok=True)  
final_df.to_csv('output/test/out.csv')  
```
